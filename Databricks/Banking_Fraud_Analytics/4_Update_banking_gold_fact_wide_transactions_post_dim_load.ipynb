```python
%sql
MERGE INTO banking_gold.fact_wide_transactions AS target
USING banking_gold.dim_customer_gold_scd1 AS source
ON target.CUST_ID = source.CUST_ID

WHEN MATCHED AND (
    target.NAME_MASKED IS DISTINCT FROM source.NAME_MASKED OR
    target.Risk_Score IS DISTINCT FROM source.Risk_Score OR
    target.KYC_Status IS DISTINCT FROM source.KYC_Status
)
THEN UPDATE SET
    target.NAME_MASKED = source.NAME_MASKED,
    target.Risk_Score = source.Risk_Score,
    target.KYC_Status = source.KYC_Status;
```

```python
%sql

MERGE INTO banking_gold.fact_wide_transactions AS target
USING banking_gold.dim_account_gold_scd1 AS source
ON target.ACCOUNT_ID = source.ACCOUNT_ID

WHEN MATCHED AND (
    target.CUST_ID IS DISTINCT FROM source.CUST_ID OR
    target.TYPE IS DISTINCT FROM source.TYPE OR
    target.BALANCE IS DISTINCT FROM source.BALANCE OR
    target.STATUS IS DISTINCT FROM source.STATUS
)
THEN UPDATE SET
    target.CUST_ID = source.CUST_ID,
    target.TYPE = source.TYPE,
    target.BALANCE = source.BALANCE,
    target.STATUS = source.STATUS;
```
