```python
%sql
CREATE TABLE IF NOT EXISTS banking_bronze.bronze_dim_account (
    ACCOUNT_ID STRING,
    CUST_ID STRING,
    TYPE STRING,
    BALANCE DECIMAL,
    STATUS STRING,
    updated_at TIMESTAMP
);
```

```python
%sql
-- Incremental load applying Watermarking Feature
INSERT INTO banking_bronze.bronze_dim_account
SELECT 
    ACCOUNT_ID, 
    CUST_ID, 
    TYPE, 
    BALANCE, 
    STATUS,  
    updated_at
FROM banking_foreign_catalog.bank_raji_schema.dim_account
WHERE updated_at > (
    SELECT COALESCE(MAX(updated_at), CAST('1900-01-01' AS TIMESTAMP)) 
    FROM banking_bronze.bronze_dim_account
);
```
